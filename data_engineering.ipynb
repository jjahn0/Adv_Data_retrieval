```python
import pandas as pd
import os
```

```python
file1 = os.path.join("Resources","hawaii_measurements.csv")
hi_data = pd.read_csv(file1)
hi_data = hi_data.dropna()
hi_data = hi_data.rename(columns={'prcp':'precipitation', 'tobs':'temperature'})
hi_data.to_csv("Resources/clean_hawaii_measurements.csv", index=False)
hi_data.head()
```

```python
file2 = os.path.join("Resources","hawaii_stations.csv")
hi_stations = pd.read_csv(file2)
hi_stations.to_csv("Resources/clean_hawaii_stations.csv", index=False)
hi_stations
```

```python
hi_station_data = pd.merge(hi_data, hi_stations, on="station", how="outer")
print (hi_station_data.count())
hi_station_data.head()
```

```python
avg_prcp_station = pd.DataFrame(hi_station_data['precipitation'].groupby(hi_station_data['station']).mean()).reset_index().rename(columns={'precipitation': 'avg_prcp'})
avg_tobs_station = pd.DataFrame(hi_station_data['temperature'].groupby(hi_station_data['station']).mean()).reset_index().rename(columns={'temperature': 'avg_tobs'})
df1 = pd.merge(avg_prcp_station, hi_stations, on='station', how='left')
avg_tobs_prcp = pd.merge(avg_tobs_station, df1, on='station', how='left')[['station', 'avg_tobs', 'avg_prcp', 'elevation']].sort_values('avg_tobs', ascending=False)
avg_tobs_prcp
```
